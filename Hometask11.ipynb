```python
import os
import shutil

menu = '''==========================================
1-Создание
2-Удаление
3-Перемещение
4-Копирование
5-Выход
==========================================
'''

menu2 = '''==========================================
1-Папка
2-Файл
3-Выход
==========================================
'''

while True:
    try:
        print(menu)
        item1 = int(input('Выберите пункт: '))
        if item1 == 1:
            while True:
                try:
                    print(menu2)
                    item2 = int(input('Выберите пункт: '))
                    if item2 == 1:
                        # Создание папки
                        print('Пример формата пути: 'r'C:\Users'r'\User'r'\Desktop'r'\folder'r'\ ')
                        dir_path = input('==========================================\nУкажите путь к папке: ')
                        name_dir = input('==========================================\nУкажите название папки: ')
                        os.mkdir(dir_path + name_dir)
                        print('Выполнено')
                    elif item2 == 2:
                        # Создание файла
                        print('Пример формата пути: 'r'C:\Users'r'\User'r'\Desktop'r'\folder'r'\ ')
                        dir_path = input('==========================================\nУкажите путь к папке: ')
                        name_file = input('==========================================\nУкажите название файла: ')
                        my_file = open(dir_path + name_file, "w+")
                        my_file.close()
                        print('Выполнено')
                    elif item2 == 3:
                        print('-== Выход ==-')
                        break
                    else:
                        print('Выберите пункт от 1 до 3')
                except ValueError:
                    print('Выберите цифру')
                except FileExistsError:
                    print('Папка с таким именем уже существует')
        elif item1 == 2:
            while True:
                try:
                    print(menu2)
                    item2 = int(input('Выберите пункт: '))
                    if item2 == 1:
                        # Удаление папки
                        print('Пример формата пути: 'r'C:\Users'r'\User'r'\Desktop'r'\folder'r'\ ')
                        dir_path = input('==========================================\nУкажите путь к папке: ')
                        name_dir = input('==========================================\nУкажите название папки: ')
                        os.rmdir(dir_path + name_dir) 
                        print('Выполнено')
                    elif item2 == 2:
                        # Удаление файла
                        print('Пример формата пути: 'r'C:\Users'r'\User'r'\Desktop'r'\folder'r'\ ')
                        dir_path = input('==========================================\nУкажите путь к папке: ')
                        name_file = input('==========================================\nУкажите название файла: ')
                        os.remove(dir_path + name_file)
                        print('Выполнено')
                    elif item2 == 3:
                        print('-== Выход ==-')
                        break
                    else:
                        print('Выберите пункт от 1 до 3')
                except ValueError:
                    print('Выберите цифру')
                except FileNotFoundError:
                    print('Папки/файла с таким именем не существует')
                except OSError:
                    print('Папка не пуста, первоначально удалите содержимое')
        elif item1 == 3:
            while True:
                try:
                    print(menu2)
                    item2 = int(input('Выберите пункт: '))
                    if item2 == 1:
                        # Перемещение папки
                        print('Пример формата пути: 'r'C:\Users'r'\User'r'\Desktop'r'\folder'r'\ ')
                        dir_path = input('==========================================\nУкажите путь к папке: ')
                        name_dir = input('==========================================\nУкажите название папки: ')
                        dir2_path = input('==========================================\nУкажите новый путь к папке: ')
                        os.replace(dir_path + name_dir, dir2_path + name_dir)
                        print('Выполнено')
                    elif item2 == 2:
                        # Перемещение файла
                        print('Пример формата пути: 'r'C:\Users'r'\User'r'\Desktop'r'\folder'r'\ ')
                        dir_path = input('==========================================\nУкажите путь к папке: ')
                        name_file = input('==========================================\nУкажите название файла: ')
                        dir2_path = input('==========================================\nУкажите новый путь к папке: ')
                        def copy_funk(dir_path, name_file, dir2_path):
                            if os.path.isfile(dir2_path + name_file) == True:
                                print('Файл с таким именем уже существует')
                            else:
                                os.replace(dir_path + name_file, dir2_path + name_file)
                                print('Выполнено')
                        copy_funk(dir_path, name_file, dir2_path)
                    elif item2 == 3:
                        print('-== Выход ==-')
                        break
                    else:
                        print('Выберите пункт от 1 до 3')
                except ValueError:
                    print('Выберите цифру')
                except FileNotFoundError:
                    print('Папки/файла с таким именем не существует')
                except PermissionError:
                    print('Папка/файл с таким именем уже существует')
        elif item1 == 4:
            while True:
                try:
                    print(menu2)
                    item2 = int(input('Выберите пункт: '))
                    if item2 == 1:
                        # Копирование папки
                        print('Пример формата пути: 'r'C:\Users'r'\User'r'\Desktop'r'\folder'r'\ ')
                        dir_path = input('==========================================\nУкажите путь к папке: ')
                        name_dir = input('==========================================\nУкажите название папки: ')
                        dir2_path = input('==========================================\nУкажите путь к папке: ')
                        name2_dir = input('==========================================\nУкажите название папки: ')
                        shutil.copytree(dir_path + name_dir, dir2_path + name2_dir)
                        print('Выполнено')
                    elif item2 == 2:
                        # Копирование файла
                        print('Пример формата пути: 'r'C:\Users'r'\User'r'\Desktop'r'\folder'r'\ ')
                        dir_path = input('==========================================\nУкажите путь к папке: ')
                        name_file = input('==========================================\nУкажите название файла: ')
                        dir2_path = input('==========================================\nУкажите путь к папке: ')
                        name2_file = input('==========================================\nУкажите новое название файла: ')
                        def copy2_funk(dir_path, name_file, dir2_path, name2_file):
                            if os.path.isfile(dir2_path + name2_file) == True:
                                print('Файл с таким именем уже существует')
                            else:
                                shutil.copy2(dir_path + name_file, dir2_path + name2_file)
                                print('Выполнено')
                        copy2_funk(dir_path, name_file, dir2_path, name2_file)
                    elif item2 == 3:
                        print('-== Выход ==-')
                        break
                    else:
                        print('Выберите пункт от 1 до 3')
                except ValueError:
                    print('Выберите цифру')
                except FileNotFoundError:
                    print('Папки/файла с таким именем не существует')
                except FileExistsError:
                    print('Папка с таким именем уже существует')
        elif item1 == 5:
            print('-== Выход ==-')
            break
        else:
            print('Выберите пункт от 1 до 5')
    except ValueError:
        print('Выберите цифру')
```
